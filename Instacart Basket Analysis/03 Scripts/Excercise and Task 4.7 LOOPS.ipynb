```python
# Defining a new function called 'price_label'
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Appling the 'price_label' on axis = 1 (rows)
df_ords_prods_merge_sub['price_range'] = df_ords_prods_merge_sub.apply(price_label, axis=1)
```

```python
# Checking the sum of each category
df_ords_prods_merge_sub['price_range'].value_counts(dropna = False)
```

2: Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.

```python
# Creating a new list and loop for the 'busiest days' and 'slowest days'
result_1 = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_1.append("Busiest days")
  elif value == 4 or value == 3:
    result_1.append("Slowest days")
  else:
    result_1.append("Regularly busy")
```

```python
result_1
```

```python
# Creating a new column in the 'df_ords_prods_merge' dataframe called 'busiest days'
df_ords_prods_merge['slowest_and_busiest_days'] = result_1
```

4: When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# Checking the frequency count of the 'order_hour_of_day' column
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Creating a new list and loop for the 'Most orders', 'Fewest orders', and 'Average orders'
result_hour = []

for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in [10,11,14,15,13,12,16,9]:
    result_hour.append("Most orders")
  elif value in [3,4,2,5,1,0,6,23]:
    result_hour.append("Fewest orders")
  else:
    result_hour.append("Average orders")
```

```python
result_hour
```

```python
# Creating the 'busiest_period_of_day' column
df_ords_prods_merge['busiest_period_of_day'] = result_hour
```

```python
# Checking the frequency count of the 'busiest_period_of_day' column
df_ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
