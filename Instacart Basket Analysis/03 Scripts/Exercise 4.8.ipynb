# 02 Importing data

```python
# Importing the 'ords_prods_merge' dataframe, 'pkl' format
ords_prods_merge = pd.read_pickle(r'/Users/peterguan/Instacart Basket Analysis/02 Data/Prepared Data/ords_prods_merge.pkl')
```

```python
# Creating a subset with ONE MILLION rows
df = ords_prods_merge[:1000000]
```

```python
# Checking the dimensions of the subset
df.shape
```

```python
# Checking the first 5 rows
df.head()
```

```python
# This function groups the dataframe by the 'product_name' column
df.groupby('product_name')
```

```python
# Using the GROUPBY and .agg (aggregate) functions
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# Using the GROUP BY function and finding the mean WITHOUT .AGG
df.groupby('department_id')['order_number'].mean()
```

```python
# Using a DOT (.) instead of square brackets [ ]
df.groupby('department_id').order_number.mean()
```

```python
# Using the GROUPBY function with MULTIPLE AGGREGATIONS
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
# Creating the 'max_order' column with all 3 STEPS in ONE LINE OF CODE
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
ords_prods_merge.head(100)
```

```python
# Creating a FLAG with '.loc' for 'loyal_customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# Creating a FLAG with '.loc' for 'regular_customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# Creating a FLAG with '.loc' for 'new customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Checking the frequency of each customer in the 'loyalty_flag' column
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```
